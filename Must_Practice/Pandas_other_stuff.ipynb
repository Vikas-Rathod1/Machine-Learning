# dropna

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                    "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                             pd.NaT]})
df
```

```python
df.dropna()
```

```python
df.dropna(axis= 1) #column (have at least single NA value)
```

```python
df.dropna(axis =0)
```

```python
df.dropna(how= 'all') #Drop the rows where all elements are missing.
```

```python
df.dropna(thresh = 2)
```

```python
## check how many col are having nan values 
#subset
df.dropna(subset = ['name','toy'])
```

# Combining dataframes

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'C': ['C0', 'C1', 'C2', 'C3'],
                     'D': ['D0', 'D1', 'D2', 'D3']},
                     index=[0, 1, 2, 3]) 

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                     'B': ['B4', 'B5', 'B6', 'B7'],
                     'C': ['C4', 'C5', 'C6', 'C7'],
                     'D': ['D4', 'D5', 'D6', 'D7']},
                      index=[4, 5, 6, 7])
 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                     'B': ['B8', 'B9', 'B10', 'B11'],
                     'C': ['C8', 'C9', 'C10', 'C11'],
                     'D': ['D8', 'D9', 'D10', 'D11']},
                     index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

# concatenation

```python
o  = pd.concat([df1,df3],axis= 1)
o
```

```python
o = pd.concat([df1,df3,df2],axis= 1)
o
```

```python
pd.concat([df1,df3,df2],axis= 1,join = 'inner')
```

# Aggregation (groupby)

```python
read_ = pd.read_csv("/home/vikas/Downloads/data/flights.csv")
rd = read_.head(12).copy()
rd
```

```python
read_.groupby('ORG_AIR')['DEP_DELAY'].count()
```

```python
read_.groupby('AIRLINE')['AIR_TIME'].sum()
```

# multple col

```python
read_.groupby('AIRLINE').agg({'ARR_DELAY':'mean'})
```

```python
rd.groupby('AIRLINE').agg({'ARR_DELAY':'mean','ARR_DELAY':'count','ARR_DELAY':'count'})
```
