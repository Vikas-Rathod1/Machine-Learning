# vectorize  Make a scalar function work on vectors

```python
def f1(x):
    if x%2==1:
        return x**2
    else:
        return x/2
f1(4)
f1(3)
```

```python
# if i want to give inputs as araay then  i will vectorieze it
f1_v = np.vectorize(f1)
f1_v
```

```python
print(f1_v([10,11,23,34]))
```

# apply_along_axis  Apply a function column wise or row wise

```python
np.random.seed(100)
```

```python
v= np.random.randint(1,150,size= (5,5))
```

```python
v
```

How to find the difference of the maximum and the minimum value in each row?

```python
v[0].max()
```

```python
def min_max(x):
    return np.max(x)-np.min(x)
np.apply_along_axis(min_max,1,v)
```
