```python
fig, (a, a1, a2) = plt.subplots(1, 3, figsize=(16,4))
fig.suptitle('One Variable Plots', size=20, y=1.02)
df.plot(kind='kde', color='green', ax=a, title='KDE plot')
df.plot(kind='box', ax=a1, title='Boxplot')
df.plot(kind='hist', color='red', ax=a2, title='Histogram')
```

```python
fig, (a, a1, a2) = plt.subplots(1, 3, figsize=(16,4))
df.sort_values('marks').plot(x='marks', y='roll_no', kind='line', ax=a)
df.sort_values('marks').plot(x='roll_no', y='marks', kind='bar', ax=a1)
df.plot(x='roll_no', kind='kde', ax=a2)
```
